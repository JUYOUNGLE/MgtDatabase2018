```python
data = pd.read_excel("2014Data.xlsx")
data.head()
```

```python
data = data.dropna(axis=1) # axis =1 열을 기준으로 삭제
data.head()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
```

```python
from sklearn import preprocessing
import pandas as pd

X=pd.DataFrame(data, columns=["X1","X2","X3","X4","X5","X6","X7","X8","X9","X10"])
Y=pd.DataFrame(data, columns=["Y1"])
```

```python
# 만족도와 재방문의사 회귀분석
linear_regression = sm.OLS(Y, X) # (종속변수, 독립변수)

# OLS 회귀모형 피팅
result_data = linear_regression.fit()

# 회귀분석 결과 요약
print(result_data.summary())
```
