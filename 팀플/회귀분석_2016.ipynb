```python
data = pd.read_excel("dd.xlsx")
data.head()
```

```python
data = data.dropna(axis=1) # axis =1 열을 기준으로 삭제
data.head()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.stats.api as sms
```

```python
from sklearn import preprocessing
import pandas as pd

X=pd.DataFrame(data, columns=["X1","X2","X3","X5","X6","X7","X8","X9","X10"])
Y=pd.DataFrame(data, columns=["Y1"])
```

```python
# 만족도와 재방문의사 회귀분석
linear_regression = sm.OLS(Y, X) # (종속변수, 독립변수)

# OLS 회귀모형 피팅
result_data = linear_regression.fit()

# 회귀분석 결과 요약
print(result_data.summary())
```

•모델의 설명력(R-squared)
◾즉, 결정계수가 0.967이라면 Y의 변동 중 96.7%가 회귀모형에 의해 설명되고 나머지 3.3%는 모형에 포함되지 않는 다른 요인들로 설명됨.
◾단순 선형회귀에서는 상관계수r의 제곱과 정확히 일치함. r^2 = R^2
◾r=0.983 -> r^2=0.967 -> R^2 = 0.967 X와 Y의 상관관계가 높을 수록 모형에 의해 데이터가 많이 설명될 수 있다.

```python
from sklearn import preprocessing
import pandas as pd

X=pd.DataFrame(data, columns=["X1","X2","X3","X5","X6","X7","X8","X9","X10"])
Y=pd.DataFrame(data, columns=["Y2"])
```

```python
# 만족도와 추천의사 회귀분석
linear_regression = sm.OLS(Y, X) # (종속변수, 독립변수)

# OLS 회귀모형 피팅
result_data = linear_regression.fit()

# 회귀분석 결과 요약
print(result_data.summary())
```
